```python
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn import svm
import numpy as np
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
```

```python
titanic.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic[titanic.Embarked.isnull()]
```

```python
titanic.drop(['Cabin'],axis=1,inplace=True)
titanic["Age"]=titanic["Age"].fillna(titanic["Age"].median())
titanic["Embarked"]=titanic["Embarked"].fillna("C")
titanic.Sex=titanic.Sex.replace(['male','female'],[0,1])
titanic.Embarked=titanic.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
features=["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
```

```python
titanic_test=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv')
titanic_test.head()
```

```python
titanic_test.isnull().sum()
```

```python
titanic_test.drop(['Cabin'],axis=1,inplace=True)
titanic_test["Age"]=titanic_test["Age"].fillna(titanic["Age"].median())
titanic_test["Embarked"]=titanic_test["Embarked"].fillna("C")
titanic_test.Sex=titanic_test.Sex.replace(['male','female'],[0,1])
titanic_test.Embarked=titanic_test.Embarked.replace(['S','C','Q'],[0,1,2])
```
