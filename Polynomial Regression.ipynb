```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
degree=2
```

```python
x_poly=np.column_stack([x**i for i in range(degree+1)])
x_poly
# column_stack-->to convert it into matrix
```

```python
weights=np.zeros(degree+1)
weights
```

```python
learning_rate=0.001
num_iterations=10000
```

```python
for _ in range(num_iterations):
  # calculate predictions
  predictions=np.dot(x_poly, weights)
  #calculte error
  error=predictions-y
  # update the weights(coefficients)using gradient descendent
  gradient=np.dot(x_poly.T,error)/len(x)
  weights -= learning_rate * gradient
  print(f'weights for iteration{_}:{weights}')
```

```python
def predict(x, weights):
  x_poly = np.array([x**i for i in range(degree + 1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(x, y , label="Data Points")
x_range = np.linspace(min(x),max(x),100)
y_pred = [predict(x,weights) for x in x_range]
plt.plot(x_range,y_pred,color='r',label="Polynomial Regression")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
